# **Project Summary -**

Mental health has become an increasingly important topic in modern workplaces, especially in the technology sector where employees often face high workloads, tight deadlines, and constant performance pressure. This project focuses on analyzing a Mental Health in Tech Survey dataset collected in 2014 to understand employees’ attitudes, experiences, and workplace support related to mental health conditions. The primary objective of this project is to explore the dataset using Exploratory Data Analysis (EDA) techniques and extract meaningful insights that can help organizations improve mental health awareness and support systems.

The dataset contains responses from individuals working in the tech industry across different countries. It includes demographic details such as age, gender, country, and employment type, along with information related to mental health history, treatment, and family background. In addition, it captures important workplace-related factors such as company size, remote work culture, availability of mental health benefits, wellness programs, anonymity protection, and ease of taking medical leave. The dataset also records employee perceptions about discussing mental health issues with coworkers, supervisors, and potential employers, as well as whether mental health is treated as seriously as physical health in their organizations.

Through EDA, the project aims to clean, preprocess, and visualize the data to identify patterns, trends, and relationships between different variables. Data cleaning steps include handling missing values, standardizing categorical variables, and removing inconsistencies. Visualization techniques such as bar charts, pie charts, histograms, heatmaps, and correlation plots are used to better understand distributions and dependencies within the dataset. These visual tools help in clearly communicating insights and making the analysis more intuitive.

One of the main goals of the project is to study how mental health awareness and support vary across geographic locations and company environments. For example, the analysis explores whether employees in certain countries are more open to seeking treatment, or whether large organizations provide better mental health benefits compared to smaller companies. The project also examines how factors like remote work, anonymity protection, and employer wellness programs influence employees’ willingness to seek help.

Another important focus is on workplace stigma. By analyzing responses related to fear of negative consequences, willingness to discuss mental health issues, and observed discrimination, the project highlights the challenges employees still face when dealing with mental health concerns in professional environments. Comparisons between attitudes toward mental and physical health further reveal whether organizations truly treat both equally.

The findings from this project are expected to provide valuable insights for employers, HR teams, and policymakers. Understanding the key predictors of mental health issues and supportive workplace practices can help organizations design better policies, improve employee well-being, and reduce stigma. For students and data analysts, this project also serves as a practical application of EDA techniques, data visualization, and storytelling with data.

The final outcome of the project is presented in the form of an EDA video, where important observations, graphs, and conclusions are explained in a simple and engaging manner. Overall, this project not only strengthens data analysis skills but also contributes to raising awareness about the importance of mental health in the tech industry and the need for more inclusive and supportive work environments.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
````

````python
# 1. Handle Missing Values
print("Handling missing values...")
# Drop 'comments' as it has too many missing values
if 'comments' in df.columns:
    df = df.drop(columns=['comments'])
````

````python
# Fill 'work_interfere' with "Don't know"
df['work_interfere'] = df['work_interfere'].fillna("Don't know")
````

````python
# Fill 'self_employed' with "No"
df['self_employed'] = df['self_employed'].fillna('No')
````

````python
# Fill 'state' with '0' (for non-US or unspecified)
df['state'] = df['state'].fillna('0')
````

````python
# 2. Standardize Gender
print("Standardizing Gender column...")
male_str = ["male", "m", "male-ish", "maile", "cis male", "mal", "male (cis)", "make", "male ", "man", "msle", "mail", "malr", "cis man", "cis male"]
female_str = ["cis female", "f", "female", "woman",  "femake", "female ", "cis-female/femme", "female (cis)", "femail"]

def clean_gender(gender):
    try:
        gender = str(gender).lower().strip()
        if gender in male_str:
            return 'Male'
        elif gender in female_str:
            return 'Female'
        else:
            return 'Trans/Other'
    except:
        return 'Trans/Other'

df['Gender'] = df['Gender'].apply(clean_gender)
````

````python
# 3. Clean Age
print("Cleaning Age column...")
# Filter ages to be between 18 and 100
df = df[(df['Age'] >= 18) & (df['Age'] <= 100)]
````

````python
# Save cleaned data
output_file = "survey_cleaned.csv"
df.to_csv(output_file, index=False)
print(f"Cleaning Complete! Cleaned data saved to '{output_file}'")
````
